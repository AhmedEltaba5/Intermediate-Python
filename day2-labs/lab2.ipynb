## heatmaps to visualize the flights dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="white") #set grid style to white
```

```python
#load dataset from seaborn datasets
flights_dataset = sns.load_dataset("flights")
flights_dataset.head() #show first five records
```

```python
# call pivot to make indexing for month and year
# pivot uses unique values from specified index / columns to form axes of flights dataframe
# index is month , column is year, and values is passengers
flights = flights_dataset.pivot("month", "year", "passengers")

# create the figure 
# figsize indicates size of the figure in inches
#returns figure and axes
f, ax = plt.subplots(figsize=(12, 6)) # Width, height in inches
sns.heatmap(flights, annot=True, fmt="d", linewidths=.5, ax=ax)
# annot parameter true to show the values in the heat map
# fmt parameter to define string format of the values
# linewidth parameter to define width of the lines that divide each cell
# ax parameter to define the axes in which will draw the plot
```

## visualize the mpg dataset

```python
# Load the example mpg dataset
mpg = sns.load_dataset("mpg")
mpg.head() #show head
```

```python
# Plot miles per gallon against horsepower with other semantics
sns.relplot(x="horsepower", y="mpg", hue="origin", size="weight",
            sizes=(40, 400), alpha=.5, palette="muted",
            height=6, data=mpg)
#this function takes the following parameters
# x , y are the variables that specify positions on the x and y axes
# hue is the grouping variable that will produce elements with different colors
# we will group by the origin country
# size is weight => so that produce elements with different sizes in grouping
# sizes is tuple specifying the minimum and maximum size to use such that other values are normalized within this range
# data is mpg our dataset
# palette is method for choosing the colors to use when mapping the hue semantic
# height is  the height (in inches) of each facet
```

### Visualize neural network output for testing image

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
#get the image
img = mpimg.imread('dog.png')
```

```python
#predictions
#first => cat , second => dog
predictions_from_neuralnetwork = np.array([4.1542e-04, 9.9958e-01])
```

```python
fig = plt.figure("neural network cat & dog classification ")

img_pos = fig.add_subplot(1 ,2, 1)
probability_pos = fig.add_subplot(1 ,2, 2)

img_pos.imshow(img)
img_pos.axis('off')

probability_pos.barh(['cat', 'dog'], predictions_from_neuralnetwork)
probability_pos.set_aspect(0.5) #size of the plot container

probability_pos.set_yticks(np.arange(2))
probability_pos.set_yticklabels(['cat', 'dog'], size='large');
                            
probability_pos.set_title("Model Output Visulaization")
probability_pos.set_xlim(0, 1.1)

plt.tight_layout()
```

# Question 2
## Stacked bar chart

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 5
menValues = (20, 35, 30, 35, 27)

ind = np.arange(N)  # the x locations for the groups
width = 0.35       # the width of the bars

#add yerr for error bars in chart
men_err = [2, 3, 4, 1, 2]
women_err = [3, 5, 2, 3, 3]

fig, ax = plt.subplots()
menBar = ax.bar(ind, menValues, width, color='b', yerr=men_err)

womenValues = (25, 32, 34, 20, 25)
womenBar = ax.bar(ind , womenValues, width, color='y', bottom=menValues, yerr=women_err)


# add some text for labels, title and axes ticks
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')

ax.set_xticks(ind)
ax.set_xticklabels(('G1', 'G2', 'G3', 'G4', 'G5'))

ax.legend((menBar, womenBar), ('Men', 'Women'))


plt.show()
```

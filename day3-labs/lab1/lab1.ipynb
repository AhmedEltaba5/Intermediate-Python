## Question 2 -- Titanic dataset

```python
titanic_csvDF = pd.read_csv('datasets/2 process_Titanic_datasets/titanic-passengers.csv', sep=';')
titanic_csvDF.head()
```

```python
titanic_jsonDF = pd.read_json('datasets/2 process_Titanic_datasets/titanic-passengers.json')
titanic_jsonDF.head()
```

```python
titanic_jsonDF.drop(['datasetid', 'recordid', 'record_timestamp'], axis = 1, inplace=True)
titanic_jsonDF.head()
```

```python
print(titanic_jsonDF.iloc[0]) #it is a dictionary!!
```

```python
# convert fileds dictionary to columns
from pandas import json_normalize
```

```python
titanic_jsonDF = json_normalize(titanic_jsonDF.fields)
```

```python
titanic_jsonDF.head()
```

```python
## remove Age & ticket from titanic_csvDF
titanic_csvDF.drop(['Age', 'Ticket'], axis = 1, inplace=True)
titanic_csvDF.head()
```

```python
## remove Fare & cabin from titanic_jsonDF
titanic_jsonDF.drop(['fare', 'cabin'], axis = 1, inplace=True)
```

```python
## combine the two dataframes inner join on pclass, name, sex, sibsp, parch, embarked
## rename columns first
titanic_csvDF.rename(columns={"Pclass": "pclass",
                              "Name": "name",
                              "Sex": "sex",
                              "SibSp": "sibsp",
                              "Parch": "parch",
                              "Embarked": "embarked",
                              "PassengerId": "passengerid",
                              "Survived": "survived"}, inplace=True)
pd.merge(titanic_csvDF, titanic_jsonDF, on=['pclass', 'name', 'sex', 'sibsp', 'parch', 'embarked'], how='inner')
```

```python
titanic_merged_df = pd.merge(titanic_csvDF, titanic_jsonDF, on=['pclass', 'name', 'sex', 'sibsp', 'parch',
                                                                'embarked', 'passengerid', 'survived'], how='inner')
titanic_merged_df.head()
```

```python
#write merged DF to csv 
titanic_merged_df.to_csv("titanic_merged.csv", index=False, sep=';')
```
